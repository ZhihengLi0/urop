# SuperCDMS LED Data Analysis

This notebook analyzes phonon detector data from LED calibration runs to verify detector stability.  
Data files: `23231213_192731_F0003.mid.gz` (400 events), `23231213_192731_F0012.mid.gz` (31 events)  
Detectors: Z1, Z2, Z3 — channels: PxS1 / PxS2 (11 total; T3Z3 has no PAS1)

**Pipeline overview**
1. Load MIDAS raw data via `rawio`
2. Filter out no-pulse and hardware-offset events
3. Align all pulses to a common time reference
4. Detect glitch events (pre-pulse negative dip)
5. Save glitch event index dictionary to disk

## 1. Setup

````python
import rawio
import numpy as np
from scipy.signal import butter, lfilter
import matplotlib.pyplot as plt
import cdms
print("CDMS Software Version", cdms.get_global_version())
````

````python
def butter_lowpass_filter(data, cutoff_frequency, sampling_rate=625000, order=4):
    # cutoff_frequency in kHz; attenuates high-frequency noise above this threshold
    cutoff_frequency *= 1000
    nyquist = 0.5 * sampling_rate
    normal_cutoff = cutoff_frequency / nyquist
    b, a = butter(order, normal_cutoff, btype='low', analog=False)
    return lfilter(b, a, data)
````

````python
# T3Z3 has no PAS1 channel (hardware absent), so it is excluded from the list
chanNames = ["PBS1", "PCS1", "PDS1", "PES1", "PFS1",
             "PAS2", "PBS2", "PCS2", "PDS2", "PES2", "PFS2"]
````

## 4. Analysis Pipeline

### 4.1 Configuration

All tunable thresholds are defined here. Change them in one place and every downstream cell picks up the update.

````python
# ==============================================================
# Analysis configuration — tune all thresholds here
# ==============================================================

FS = 625000  # detector sampling rate (Hz)

# --- Event filtering ---
MIN_AMPLITUDE = 2000   # min peak-to-peak (ADC counts) to count as a pulse
                       # events below this are noise or empty triggers
MAX_BASELINE  = 33100  # max |baseline| (ADC counts); event12 in F0012 sits at ~33200,
                       # indicating a hardware baseline offset — excluded
BASELINE_N    = 5000   # samples at trace start used to estimate baseline

# --- Pulse alignment ---
GAP_MS         = 0.05  # time window (ms) over which the rising-edge slope is measured
                       # at 625 kHz this is ~31 samples
RISE_THRESHOLD = 0.2   # normalized slope threshold that identifies the pulse leading edge
ALIGN_MS       = 25.3  # target time (ms) where every pulse's rising edge is placed

# --- Glitch detection ---
# A glitch appears as a downward dip just *before* the LED pulse
GLITCH_WIN = (24.9, 25.25)  # time window (ms) to check for pre-pulse dip
GLITCH_THR = -0.05          # normalized amplitude threshold; dip below this → glitch
````

### 4.2 Helper Functions

````python
def find_and_align(y_raw, x0):
    """
    Filter, align, and normalize a single phonon channel trace.

    Steps:
      1. Reject events with no pulse or with a hardware baseline offset.
      2. Subtract baseline; normalize peak to 1.
      3. Find the pulse rising edge with a vectorized slope scan
         (replaces the old per-step while-loop for a ~100x speed-up).
      4. Shift the time axis so the rising edge lands at ALIGN_MS.

    Returns
    -------
    (x_aligned, y_norm) : tuple of np.ndarray
        Aligned time axis and normalized trace, or (None, None) if rejected.
    """
    y = y_raw.astype(np.int32)
    baseline = np.mean(y[:BASELINE_N])

    # Reject: no clear pulse or hardware baseline offset
    if (max(y) - min(y)) <= MIN_AMPLITUDE or abs(baseline) >= MAX_BASELINE:
        return None, None

    y = y - baseline
    peak = float(np.max(y))
    if peak == 0:  # guard against flat / inverted trace
        return None, None
    y = y / peak  # normalize so pulse maximum = 1

    # Find rising edge: compute slope = y[i + gap] - y[i] at every sample
    # This is equivalent to the old while-loop scan but done as one numpy operation
    gap_samples = int(GAP_MS * 1e-3 * FS)   # ms → samples (~31 at 625 kHz)
    search_end  = np.searchsorted(x0, 29.0)  # don't scan past 29 ms

    slope      = y[gap_samples:search_end] - y[:search_end - gap_samples]
    candidates = np.where(slope > RISE_THRESHOLD)[0]

    if len(candidates) == 0:  # no clear rising edge found
        return None, None

    rise_idx  = candidates[0]  # index of first sample where slope exceeds threshold
    x_aligned = x0 - (x0[rise_idx] - ALIGN_MS)  # shift time axis
    return x_aligned, y


def is_glitch(x, y):
    """
    Return True if the aligned trace has a significant negative dip
    in the pre-pulse window — the signature of a glitch event.
    """
    mask = (x >= GLITCH_WIN[0]) & (x <= GLITCH_WIN[1])
    if not np.any(mask):
        return False
    return bool(np.min(y[mask]) < GLITCH_THR)
````

### 4.3 Aligned Pulses — All Detectors and Channels

````python
# Plot all events that pass the pulse + baseline cuts, aligned and normalized
for Z_choice in ['Z1', 'Z2', 'Z3']:
    for i_det in chanNames:
        x0 = np.arange(len(events[0][Z_choice][i_det])) / FS * 1000

        n_plotted = 0
        for i, ev in enumerate(events):
            x, y = find_and_align(ev[Z_choice][i_det], x0)
            if x is not None:
                plt.plot(x, y, alpha=0.6, label=f"event {i}")
                n_plotted += 1

        if n_plotted > 0:
            plt.legend(fontsize=7)
            plt.xlabel("Time [ms]")
            plt.ylabel("Normalized amplitude")
            plt.title(f"{Z_choice} {i_det}  ({n_plotted} events aligned)")
            plt.xlim(22, 30)
            plt.show()
        else:
            plt.close()
````

### 4.4 Glitch Events — All Detectors and Channels

````python
# Plot only events flagged as glitches (pre-pulse negative dip)
for Z_choice in ['Z1', 'Z2', 'Z3']:
    for i_det in chanNames:
        x0 = np.arange(len(events[0][Z_choice][i_det])) / FS * 1000

        n_plotted = 0
        for i, ev in enumerate(events):
            x, y = find_and_align(ev[Z_choice][i_det], x0)
            if x is not None and is_glitch(x, y):
                plt.plot(x, y, label=f"event {i}")
                n_plotted += 1

        if n_plotted > 0:
            plt.legend()
            plt.xlabel("Time [ms]")
            plt.ylabel("Normalized amplitude")
            plt.title(f"{Z_choice} {i_det} — glitch events only")
            plt.xlim(22, 30)
            plt.show()
        else:
            plt.close()
````

## 5. Build Glitch Event Dictionary

Collect glitch event indices into a nested dictionary and save to disk so other notebooks can load it without re-running the full analysis.

```
glitch_event_dic[detector][channel] = [list of event indices]
```

````python
glitch_event_dic = {}

for Z_choice in ['Z1', 'Z2', 'Z3']:
    glitch_event_dic[Z_choice] = {}
    for i_det in chanNames:
        glitch_event_dic[Z_choice][i_det] = []
        x0 = np.arange(len(events[0][Z_choice][i_det])) / FS * 1000

        for i, ev in enumerate(events):
            x, y = find_and_align(ev[Z_choice][i_det], x0)
            if x is not None and is_glitch(x, y):
                glitch_event_dic[Z_choice][i_det].append(i)

save_path = "/home/yanliusp/Shared/Zhiheng/glitch_event_dic.npy"
np.save(save_path, glitch_event_dic)
print("Saved to:", save_path)
print(glitch_event_dic)
````
