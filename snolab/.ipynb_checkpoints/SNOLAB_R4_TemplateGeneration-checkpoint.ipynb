```python
# ── Paths and global config ────────────────────────────────────────────────────
PROCESSED_DIR = "/projects/standard/yanliusp/shared/data/CDMS/SNOLAB/R4/Processed/Prompt/Prompt_V07-02_C0.4.5/Submerged"
RAW_DIR       = "/projects/standard/yanliusp/shared/data/CDMS/SNOLAB/R4/Raw"
PROD_TAG      = "Prompt_V07-02_C0.4.5"
samplerate    = 625000

# Build series list from filenames
all_proc_files = sorted(glob.glob(f"{PROCESSED_DIR}/{PROD_TAG}_*.root"))
series_list    = [os.path.basename(f).replace(f"{PROD_TAG}_", "").replace(".root", "") for f in all_proc_files]
print(f"Found {len(series_list)} series:")
for s in series_list:
    print(f"  {s}")

# All detectors present in the processed files
dets_all  = [1, 4, 6, 7, 9, 10, 15, 16, 18, 24]

# Primary detector for template generation (change as needed)
det_tmpl   = 1

# Standard 12 phonon channels
chan_names = ['PAS1','PBS1','PCS1','PDS1','PES1','PFS1',
              'PAS2','PBS2','PCS2','PDS2','PES2','PFS2']
```

## Part 2 — 1×1 Template Generation

**Pipeline:**  
1. CDataFrame event selection (TriggerType==1, PTOFamps range)  
2. Read raw traces via `rawio`  
3. Baseline subtract → Butterworth low-pass filter → align with `OFdelay`  
4. Average normalised traces → template  
5. Fit (3-exp preferred, 2-exp fallback)  
6. Build PT / PS1 / PS2 composites  
7. Write to ROOT file  

### Step 1 — Event Selection with CDataFrame

```python
# PTOFamps selection window — inspect the spectrum above and adjust
# These are loose bounds to keep signal events and reject noise / saturation
PTOF_LO = 1e-8   # lower bound  [A]
PTOF_HI = 1e-5   # upper bound  [A]

# Collect per-series: event numbers, OFdelay, and bs for each channel
sel_events   = {}   # sel_events[series]   = 1-D array of EventNumber
sel_delay    = {}   # sel_delay[series][chan]  = dict {EventNumber: delay_s}
sel_baseline = {}   # sel_baseline[series][chan] = dict {EventNumber: baseline_ADC}

for fpath, series in zip(all_proc_files, series_list):
    try:
        df = CDataFrame(
            f"rqDir/zip{det_tmpl}",
            fpath,
            [f"{fpath}:rqDir/eventTree"]
        )
        df = df.Filters([
            "TriggerType==1",
            f"PTOFamps>{PTOF_LO}",
            f"PTOFamps<{PTOF_HI}"
        ])

        cols = ["EventNumber"] + [f"{c}OFdelay" for c in chan_names] + [f"{c}bs" for c in chan_names]
        data = df.AsNumpy(columns=cols)

        evnums = data["EventNumber"]
        sel_events[series] = evnums
        sel_delay[series]    = {}
        sel_baseline[series] = {}
        for c in chan_names:
            sel_delay[series][c]    = {int(evnums[i]): data[f"{c}OFdelay"][i] for i in range(len(evnums))}
            sel_baseline[series][c] = {int(evnums[i]): data[f"{c}bs"][i]      for i in range(len(evnums))}

        print(f"  {series}: {len(evnums)} events selected")
    except Exception as e:
        print(f"  {series}: ERROR — {e}")
        sel_events[series]   = np.array([], dtype=int)
        sel_delay[series]    = {c: {} for c in chan_names}
        sel_baseline[series] = {c: {} for c in chan_names}

total_sel = sum(len(v) for v in sel_events.values())
print(f"\nTotal selected events for Zip{det_tmpl}: {total_sel}")
```

### Step 2 — Read Raw Traces, Filter and Align

```python
def butter_lowpass(data, cutoff_khz=10.0, fs=625000, order=4):
    sos = butter(order, cutoff_khz * 1000, btype='low', fs=fs, output='sos')
    return sosfilt(sos, data)


def preprocess(y_raw, snr_thresh=5.0, baseline_n=5000):
    """Baseline-subtract, reject noise events by SNR.
    Returns (y_sub, ok) where y_sub is baseline-subtracted ADC."""
    y = y_raw.astype(np.float64)
    baseline = np.mean(y[:baseline_n])
    y = y - baseline
    peak = np.max(y)
    if peak <= 0:
        return None, False
    noise_std = np.std(y[:baseline_n])
    if noise_std <= 0 or peak / noise_std < snr_thresh:
        return None, False
    return y, True
```

```python
# Read raw traces and build aligned, normalised traces per channel
# channel_traces[chan] = list of normalised aligned traces
channel_traces = {c: [] for c in chan_names}

for series, evnums in sel_events.items():
    if len(evnums) == 0:
        continue

    raw_series_dir = f"{RAW_DIR}/{series}"
    if not os.path.isdir(raw_series_dir):
        print(f"  {series}: raw directory not found, skipping")
        continue

    try:
        myreader = rawio.RawDataReader(filepath=raw_series_dir, series=series)
    except Exception:
        # Fallback: pass directory without series keyword
        try:
            myreader = rawio.RawDataReader(raw_series_dir)
        except Exception as e:
            print(f"  {series}: cannot open rawio reader — {e}")
            continue

    try:
        raw_evts = myreader.read_events(
            output_format  = 1,
            skip_empty     = True,
            trigger_types  = [1],
            event_nums     = [int(n) for n in evnums],
            detector_nums  = [det_tmpl],
            channel_names  = chan_names
        )
    except Exception as e:
        print(f"  {series}: read_events failed — {e}")
        continue

    z_key = f"Z{det_tmpl}"
    n_good = 0
    for evn in evnums:
        evn = int(evn)
        for chan in chan_names:
            try:
                pulse = raw_evts[z_key][chan][series][evn][:]
            except (KeyError, IndexError, TypeError):
                continue

            y_sub, ok = preprocess(pulse)
            if not ok:
                continue

            # Low-pass filter (10 kHz)
            y_filt = butter_lowpass(y_sub)

            # Align with OFdelay (from processed RQs)
            delay_s = sel_delay[series][chan].get(evn, 0.0)
            y_align = np.roll(y_filt, -round(delay_s * samplerate))

            # Normalise to unit peak
            peak = np.max(y_align)
            if peak <= 0:
                continue
            channel_traces[chan].append(y_align / peak)
            n_good += 1

    print(f"  {series}: {n_good} good channel-trace pairs")

for c in chan_names:
    print(f"  {c}: {len(channel_traces[c])} traces total")
```

```python
# Quick-look: plot overlay of aligned traces per channel (up to 50 per panel)
sos_plot = butter(1, 10000, 'low', fs=samplerate, output='sos')
tracelength = None  # will be set from first trace

fig, axes = plt.subplots(4, 3, figsize=(14, 12), sharex=False)
axes = axes.flatten()

for idx, chan in enumerate(chan_names):
    ax = axes[idx]
    traces = channel_traces[chan]
    if tracelength is None and len(traces) > 0:
        tracelength = len(traces[0])
    for tr in traces[:50]:
        ax.plot(tr, alpha=0.2, color='black', linewidth=0.4)
    if len(traces) > 0:
        avg = np.mean(traces, axis=0)
        ax.plot(avg, color='steelblue', linewidth=1.2, label=f'avg N={len(traces)}')
    ax.set_title(chan, fontsize=10)
    ax.legend(fontsize=7)
    ax.ticklabel_format(style='sci', scilimits=(0,0))
    # zoom to pulse region — adjust if trigger position differs
    center = int(tracelength / 2) if tracelength else 16384
    ax.set_xlim(center - 2000, center + 4000)

fig.suptitle(f"Zip{det_tmpl} — Aligned normalised traces (max 50 shown, blue = average)",
             fontsize=12)
plt.tight_layout()
plt.show()
```

### Step 3 — Compute Average Template

```python
# average_trace[chan] = normalised average trace (unit-peak)
average_trace = {}
for chan in chan_names:
    traces = channel_traces[chan]
    if len(traces) == 0:
        average_trace[chan] = None
        print(f"  {chan}: no good traces")
    else:
        avg = np.mean(traces, axis=0)
        avg /= np.max(avg)
        average_trace[chan] = avg
        print(f"  {chan}: template from {len(traces)} traces")
```

### Step 4 — Exponential Fitting

```python
def two_exp_fit(x, amp1, t1, t2, baseline, pretrigger):
    return np.where(
        x <= pretrigger, baseline,
        -(amp1 * np.exp(-(x - pretrigger) / t1 / samplerate)
          - amp1 * np.exp(-(x - pretrigger) / t2 / samplerate)) + baseline
    )


def three_exp_fit(x, amp1, amp2, t1, t2, t3, baseline, pretrigger):
    return np.where(
        x <= pretrigger, baseline,
        -((amp1 + amp2) * np.exp(-(x - pretrigger) / t1 / samplerate)
          - amp1          * np.exp(-(x - pretrigger) / t2 / samplerate)
          - amp2          * np.exp(-(x - pretrigger) / t3 / samplerate)) + baseline
    )


def auto_guess(tmpl, win_start):
    """Estimate initial parameters from template shape."""
    peak_idx = int(np.argmax(tmpl))
    above5   = np.where(tmpl[win_start:peak_idx] >= 0.05)[0]
    pt_guess = float(win_start + above5[0]) - 50 if len(above5) > 0 else float(peak_idx) - 200
    pt_guess = max(float(win_start), pt_guess)
    post     = tmpl[peak_idx:]
    inv_e    = np.where(post <= 1.0 / np.e)[0]
    t2_g     = inv_e[0] / samplerate if len(inv_e) > 0 else 5e-3
    return pt_guess, t2_g * 0.05, t2_g, t2_g * 5.0
```

```python
# Fit each channel template — try 3-exp first, fall back to 2-exp
WIN_START = tracelength // 2 - 3000 if tracelength else 13384

fit_params   = {}   # fit_params[chan]   = ('3-exp'|'2-exp', popt)
fit_trace    = {}   # fit_trace[chan]    = fitted array (same length as template)
_plot_data   = {}   # for plotting

for chan in chan_names:
    raw_tmpl = average_trace[chan]
    if raw_tmpl is None:
        fit_params[chan] = None
        fit_trace[chan]  = None
        _plot_data[chan] = None
        continue

    tmpl = raw_tmpl / np.max(raw_tmpl)
    x    = np.arange(WIN_START, len(tmpl), dtype=float)
    y    = tmpl[WIN_START:]
    pt_g, t1_g, t2_g, t3_g = auto_guess(tmpl, WIN_START)

    fit_ok = False
    for try3 in [True, False]:
        try:
            with warnings.catch_warnings():
                warnings.simplefilter('ignore')
                if try3:
                    popt, _ = curve_fit(
                        three_exp_fit, x, y,
                        p0=[0.5, 0.5, t1_g, t2_g, t3_g, 0.0, pt_g],
                        maxfev=int(1e5)
                    )
                    tag   = '3-exp'
                    y_fit = three_exp_fit(x, *popt)
                    amp1, amp2, t1, t2, t3, bl, pt = popt
                    label = f't1={t1*1e3:.3f}  t2={t2*1e3:.2f}  t3={t3*1e3:.2f} ms'
                else:
                    popt, _ = curve_fit(
                        two_exp_fit, x, y,
                        p0=[1.0, t1_g, t2_g, 0.0, pt_g],
                        maxfev=int(1e5)
                    )
                    tag   = '2-exp'
                    y_fit = two_exp_fit(x, *popt)
                    amp1, t1, t2, bl, pt = popt
                    label = f't1={t1*1e3:.3f}  t2={t2*1e3:.2f} ms'

            fit_params[chan] = (tag, popt)
            # Extend fit to full trace length
            full_fit = np.full(len(tmpl), popt[-2])   # fill with baseline
            full_fit[WIN_START:] = y_fit
            fit_trace[chan] = full_fit / np.max(full_fit)

            t_ms = x / samplerate * 1000
            _plot_data[chan] = (t_ms, y, y_fit, tag, t1 * 1e3, t2 * 1e3, label)
            print(f'  {chan}: [{tag}] {label}')
            fit_ok = True
            break
        except RuntimeError:
            if try3:
                print(f'  {chan}: 3-exp failed, trying 2-exp')
            else:
                print(f'  {chan}: both fits failed')
                fit_params[chan] = None
                fit_trace[chan]  = None
                _plot_data[chan] = None
```

```python
# Plot templates with fits
fig, axes = plt.subplots(4, 3, figsize=(14, 12), sharex=False)
axes = axes.flatten()

for idx, chan in enumerate(chan_names):
    ax   = axes[idx]
    tmpl = average_trace[chan]
    ft   = fit_trace[chan]
    pd   = _plot_data[chan]

    if tmpl is not None:
        t_ms = np.arange(len(tmpl)) / samplerate * 1000
        ax.plot(t_ms, tmpl, color='steelblue', lw=1, label='avg template')
    if ft is not None and pd is not None:
        t_ms = np.arange(len(ft)) / samplerate * 1000
        ax.plot(t_ms, ft, color='red', lw=1.5, ls='--', alpha=0.85, label='fit')
        ax.text(0.97, 0.97, pd[6], ha='right', va='top',
                transform=ax.transAxes, fontsize=5.5, color='darkred')
        ax.set_title(f"{chan} [{pd[3]}]", fontsize=9)
    else:
        ax.text(0.5, 0.5, 'N/A', ha='center', va='center', transform=ax.transAxes, color='gray')
        ax.set_title(chan, fontsize=9)

    ax.legend(fontsize=7)
    ax.tick_params(labelsize=7)
    # zoom to pulse + decay
    center_ms = (tracelength // 2) / samplerate * 1000 if tracelength else 26.2
    ax.set_xlim(center_ms - 2, center_ms + 20)
    ax.set_xlabel('time [ms]', fontsize=8)

fig.suptitle(f"Zip{det_tmpl} — Templates (blue) with exponential fits (red dashed)", fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# Summary table: fitted time constants
print(f"{'Channel':8s}  {'Model':6s}  {'t1-rise [ms]':>14s}  {'t2-decay [ms]':>14s}  {'t3-decay [ms]':>14s}")
t2_vals = []
for chan in chan_names:
    entry = fit_params.get(chan)
    if entry is None:
        print(f"  {chan:<6s}:  N/A")
        t2_vals.append(0)
    elif entry[0] == '3-exp':
        _, amp2, t1, t2, t3, bl, pt = entry[1]
        print(f"  {chan:<6s}:  3-exp  {t1*1e3:>14.3f}  {t2*1e3:>14.2f}  {t3*1e3:>14.2f}")
        t2_vals.append(t2 * 1e3)
    else:
        amp1, t1, t2, bl, pt = entry[1]
        print(f"  {chan:<6s}:  2-exp  {t1*1e3:>14.3f}  {t2*1e3:>14.2f}  {'N/A':>14s}")
        t2_vals.append(t2 * 1e3)

fig, ax = plt.subplots(figsize=(10, 4))
colors = ['steelblue' if fit_params.get(c) and fit_params[c][0]=='3-exp'
          else ('orange' if fit_params.get(c) else 'lightgray') for c in chan_names]
bars = ax.bar(chan_names, t2_vals, color=colors)
for bar, v in zip(bars, t2_vals):
    if v > 0:
        ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.02,
                f'{v:.2f}', ha='center', va='bottom', fontsize=8)
ax.set_ylabel('t2-decay [ms]')
ax.set_title(f'Zip{det_tmpl} — phonon decay time constants (t2)\nblue=3-exp  orange=2-exp')
ax.tick_params(axis='x', rotation=45)
plt.tight_layout()
plt.show()
```

### Step 5 — Build PT / PS1 / PS2 and Write 1×1 Templates to ROOT

```python
# PT = mean of all 12 channels
# PS1 = mean of inner ring S1 (PAS1…PFS1)
# PS2 = mean of outer ring S2 (PAS2…PFS2)
def build_composite(chans, src_dict):
    valid = [src_dict[c] for c in chans if src_dict.get(c) is not None]
    if not valid:
        return None
    comp = np.mean(valid, axis=0)
    return comp / np.max(comp)

fit_trace['PT']  = build_composite(chan_names, fit_trace)
fit_trace['PS1'] = build_composite(chan_names[:6], fit_trace)
fit_trace['PS2'] = build_composite(chan_names[6:], fit_trace)

write_channels = chan_names + ['PT', 'PS1', 'PS2']

fig, axes = plt.subplots(1, 3, figsize=(14, 4))
for ax, label in zip(axes, ['PT', 'PS1', 'PS2']):
    tr = fit_trace.get(label)
    if tr is not None:
        t_ms = np.arange(len(tr)) / samplerate * 1000
        ax.plot(t_ms, tr, color='steelblue')
        center_ms = (tracelength // 2) / samplerate * 1000 if tracelength else 26.2
        ax.set_xlim(center_ms - 2, center_ms + 20)
    ax.set_title(f"{label} composite template", fontsize=11)
    ax.set_xlabel('time [ms]')
plt.tight_layout()
plt.show()
```

```python
# Write 1x1 templates to ROOT
output_1x1 = f"Templates_SNOLAB_R4_zip{det_tmpl}_1x1.root"
root_file  = TFile(output_1x1, "RECREATE")
directory  = root_file.mkdir(f"zip{det_tmpl}")
directory.cd()

n_bins = tracelength if tracelength else 32768

for chan in write_channels:
    tr = fit_trace.get(chan)
    if tr is None:
        print(f"  {chan}: skipped (no template)")
        continue
    tr_norm = tr / np.max(tr)
    h = TH1D(chan, chan, n_bins, 0, n_bins)
    for i, v in enumerate(tr_norm):
        h.SetBinContent(i + 1, v)
    h.Write()
    print(f"  {chan}: written ({n_bins} bins)")

root_file.Close()
print(f"\nSaved: {os.path.abspath(output_1x1)}")
```

## Part 3 — N×M Template Generation (PCA)

**Pipeline:**  
1. Re-use aligned traces; cut to a 6000-sample window around the pulse  
2. PCA to extract principal components (N_COMPONENTS per channel)  
3. Optional: remove spurious sine artefact from a component  
4. Pad components back to full trace length and write to ROOT  

```python
# Build NxM training set
# Window: [WIN_NXM_LO : WIN_NXM_HI] samples around the pulse (6000 samples)
WIN_NXM_LO   = tracelength // 2 - 3000 if tracelength else 13384
WIN_NXM_HI   = WIN_NXM_LO + 6000
TRIGGER_IDX  = 3000   # expected peak position within the 6000-sample window
N_COMPONENTS = 4      # number of PCA components per channel
MAX_EVENTS   = 200    # max training events per channel (increase if more available)

training_set = {c: [] for c in chan_names}  # training_set[chan] = list of cut traces

for chan in chan_names:
    for tr in channel_traces[chan][:MAX_EVENTS]:
        cut = tr[WIN_NXM_LO:WIN_NXM_HI]
        if len(cut) < 6000:
            continue
        # Additional alignment check: peak should be within window
        peak_pos = int(np.argmax(cut))
        if peak_pos < 2000 or peak_pos > 4000:
            continue
        training_set[chan].append(cut)
    print(f"  {chan}: {len(training_set[chan])} training events")
```

```python
# Plot training set overlay for a few channels
preview_chans = ['PBS1', 'PCS1', 'PBS2', 'PCS2']
fig, axes = plt.subplots(1, len(preview_chans), figsize=(14, 4))
for ax, chan in zip(axes, preview_chans):
    for tr in training_set[chan][:50]:
        ax.plot(tr, alpha=0.2, color='black', linewidth=0.4)
    if training_set[chan]:
        ax.plot(np.mean(training_set[chan], axis=0), color='steelblue', linewidth=1.5)
    ax.set_title(chan)
    ax.ticklabel_format(style='sci', scilimits=(0,0))
plt.suptitle('NxM training set — aligned cut traces (6000 samples)', fontsize=11)
plt.tight_layout()
plt.show()
```

```python
# PCA — fit on combined training set (sum over channels, all events)
# Build a single combined matrix for the detector
combined = []
for chan in chan_names:
    if training_set[chan]:
        combined.extend(training_set[chan][:MAX_EVENTS])

if len(combined) == 0:
    raise RuntimeError("No training events available for PCA — check event selection")

pca = PCA(N_COMPONENTS, svd_solver='full').fit(combined)
PC  = pca.components_          # shape (N_COMPONENTS, 6000)
var = pca.explained_variance_ratio_

print(f"PCA on {len(combined)} traces ({N_COMPONENTS} components)")
for i, v in enumerate(var):
    print(f"  PC{i}: {v*100:.2f}% variance")

# Variance plot
fig, ax = plt.subplots(figsize=(6, 4))
ax.plot(range(N_COMPONENTS), var, 'o-', color='steelblue')
ax.set_xlabel('Component index')
ax.set_ylabel('Explained variance ratio')
ax.set_title(f'Zip{det_tmpl} PCA explained variance')
ax.set_yscale('log')
ax.grid(alpha=0.4)
plt.tight_layout()
plt.show()
```

```python
# Plot PCA components
fig, axes = plt.subplots(2, (N_COMPONENTS + 1) // 2, figsize=(14, 6))
axes = axes.flatten()
x_cut = np.arange(6000)
for i in range(N_COMPONENTS):
    ax = axes[i]
    ax.plot(x_cut, PC[i], linewidth=0.8)
    ax.set_title(f'PC{i}  ({var[i]*100:.2f}%)', fontsize=10)
    ax.ticklabel_format(style='sci', scilimits=(0,0))
for i in range(N_COMPONENTS, len(axes)):
    axes[i].set_visible(False)
fig.suptitle(f'Zip{det_tmpl} — PCA components (N×M templates)', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# Write NxM templates to ROOT file
# Each component is padded back to full trace length (zeros at start and end)
# One histogram per channel per component: e.g. PAS1nxm0, PAS1nxm1 ...
output_nxm = f"Templates_SNOLAB_R4_zip{det_tmpl}_1x1.root"  # append to same file
rootfile   = ROOT.TFile.Open(output_nxm, 'UPDATE')
rootfile.cd(f"zip{det_tmpl}")

n_bins = tracelength if tracelength else 32768
n_pre  = WIN_NXM_LO   # samples before the cut window
n_post = n_bins - WIN_NXM_HI  # samples after the cut window

for chan in chan_names:
    for i in range(N_COMPONENTS):
        hname = f"{chan}nxm{i}"
        h = ROOT.TH1D(hname, hname, n_bins, 0, n_bins)
        padded = np.concatenate([
            np.zeros(n_pre),
            PC[i],
            np.zeros(n_post)
        ])
        # Safety: trim to n_bins if rounding mismatch
        padded = padded[:n_bins]
        peak_p = np.max(np.abs(padded))
        y = padded / peak_p if peak_p > 0 else padded
        for j, v in enumerate(y):
            h.SetBinContent(j + 1, v)
        h.Write()
        h.Delete()

rootfile.Close()
print(f"NxM templates ({N_COMPONENTS} components × {len(chan_names)} channels) written.")
print(f"Saved: {os.path.abspath(output_nxm)}")
```
